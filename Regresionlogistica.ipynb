```python
import pandas as pd

# Cargar el dataset
df = pd.read_csv("framingham.csv")  # asegúrate que el archivo esté en la misma carpeta
df = df.dropna()  # quitar filas con datos faltantes
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Separar variables y etiqueta
X = df.drop(columns=["TenYearCHD"])
y = df["TenYearCHD"]

# Escalar características
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

df.head()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Modelo de regresión logística con Keras
model = Sequential([
    Dense(1, activation='sigmoid', input_shape=(X_train.shape[1],))
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Entrenar
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=1)

# Aviso de finalización
print("Modelo entrenado correctamente. Puedes continuar con la evaluación.")
```

```python
# Evaluar
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Precisión en test: {accuracy:.4f}")

# Predicciones
predicciones = model.predict(X_test)
predicciones_bin = (predicciones > 0.5).astype(int)

# Mostrar algunas predicciones reales vs predichas
for real, pred in zip(y_test.values[:10], predicciones_bin[:10]):
    print(f"Real: {real}, Predicho: {pred[0]}")
```
